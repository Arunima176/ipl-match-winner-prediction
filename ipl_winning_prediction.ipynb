```python
df_1=df_1.drop(columns=['id','date'])
```

```python
categorical_columns=['Season', 'city','team1', 'team2', 'toss_winner','toss_decision', 'result','player_of_match', 'venue', 'umpire1','umpire2','winner']
```

```python
le=LabelEncoder()
```

```python
for col in categorical_columns:
  df_1[col]=le.fit_transform(df_1[col])

```

```python
df_1.head()
```

```python
x=df_1.drop('winner',axis=1)
```

```python
y=df_1['winner']
```

Normalization

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(n_jobs=4)
}
```

```python
results = {}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)

    results[name] = {
        'Accuracy': acc,
        'Classification Report': report
    }

for model_name, metrics in results.items():
    print(f"\n{model_name} Results:")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
    print(f"\nClassification Report:\n{metrics['Classification Report']}")
```

Hyperparameter Tuning

```python
dtc_model=models['Decision Tree']
```

```python
params_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [5, 10, 15, 20],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_model=GridSearchCV(estimator=dtc_model,param_grid=params_grid,cv=6,verbose=3,scoring='accuracy')
```

```python
grid_model.fit(x_train,y_train)
```

```python
grid_model.best_score_
```

```python
best_params=grid_model.best_params_
best_params
```

```python
best_model = grid_model.best_estimator_
```

```python
y_pred1=best_model.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred1))
```

```python
print(classification_report(y_test, y_pred1))
```
